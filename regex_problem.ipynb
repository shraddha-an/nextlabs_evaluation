```python
# String copied from the evaluation doc.
regex_string = '{"orders":[{"id":1},{"id":2},{"id":3},{"id":4},{"id":5},{"id":6},{"id":7},{"id":8},{"id":9},{"id":10},{"id":11},{"id":648},{"id":649},{"id":650},{"id":651},{"id":652},{"id":653}],"errors":[{"code":3,"message":"[PHP Warning #2] count(): Parameter must be an array or an object that implements Countable (153)"}]}'
```

# **2) Regex Extraction**

Screenshot 2022-08-27 at 10.17.10.png

```python
only_digits = re.findall(r'\:(\d+)', regex_string)
only_digits
```

```python
# Converting the strings to ints
numbers = [int(x) for x in only_digits]
numbers
```
